```python
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
df_20_21 = pd.read_csv('/content/drive/MyDrive/Data project folder/Colonial Student File 21.csv')
df_20_21.head()
```

```python
df_20_21.info()
```

```python
df_20_21['Academic Year'] = '2020-2021'
df_20_21['District'] = 'COLONIAL'

df_20_21=df_20_21[['Student ID', 'Academic Year',
                   'District', 'School',
                   'Grade', 'Race',
                   'ELA Teacher ID', 'ELA Score', 'ELA Proficiency Level',
                   'Math Teacher ID', 'Math Score','Math Proficiency Level']]
df_20_21=df_20_21.rename(columns={'Race':'Race/Ethnicity'})
```

```python
df_20_21.loc[df_20_21['Student ID'].duplicated()]
```

```python
#select the columns with ela score and proficiency.
df_20_21_ela=df_20_21[[
    'Student ID','Academic Year','District', 'Grade','Race/Ethnicity',
    'ELA Teacher ID','ELA Score','ELA Proficiency Level','School']]
#add column
df_20_21_ela['Course']='ELA'
#drop null values
df_20_21_ela = df_20_21_ela.dropna(
    how='any',
    subset=['ELA Teacher ID', 'ELA Score', 'ELA Proficiency Level'])
df_20_21_ela=df_20_21_ela.rename(
    columns={'ELA Teacher ID':'Teacher ID',
             'ELA Score':'Score', 'ELA Proficiency Level':'Proficiency Level'})
#change the type of score from float to integer
df_20_21_ela['Score'] = df_20_21_ela['Score'].astype('int')
```

```python
df_20_21_math=df_20_21[[
    'Student ID','Academic Year','District', 'Grade','Race/Ethnicity',
    'Math Teacher ID','Math Score','Math Proficiency Level','School']]
df_20_21_math['Course']='MATH'
df_20_21_math = df_20_21_math.dropna(
    how='any',
    subset=['Math Teacher ID', 'Math Score', 'Math Proficiency Level'])
df_20_21_math=df_20_21_math.rename(
    columns={'Math Teacher ID':'Teacher ID',
             'Math Score':'Score',
             'Math Proficiency Level':'Proficiency Level'})

df_20_21_math['Score'] = df_20_21_math['Score'].astype('int')
```

```python
#append both ela and maths score
df_20_21 = pd.concat([df_20_21_ela, df_20_21_math], axis=0)
```

```python

df_c1 = (df_20_21.groupby('Student ID')['District'].count().reset_index())
assert len(df_c1[df_c1['District']>2]) == 0
```

```python
df_20_21.to_csv('/content/drive/MyDrive/Data project folder/Colonial Student File 21_clean.csv',
                index=False, encoding='utf-8')
```
