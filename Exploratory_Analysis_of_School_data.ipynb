Cleaning a student score data :
-create a year column
-re-arrange column values
-create a data with student id with unique teacher id for each subject.

```python
#import pandas
import pandas as pd
```

```python
#Import the excel file
df=pd.read_excel('/content/21-22-student.xlsx')
df.head()
```

```python
# Add new columns
df[['Academic Year']]='2020-21'
df[['District']]='Rio Grande'
```

```python
# Select the columns
df1=df[['Student ID','Academic Year','District', 'Grade','Race/Ethnicity','ELA Teacher ID', 'Math Teacher ID','ELA Score','ELA Proficiency Level',
'Math Score','Math Proficiency Level','School']]
```

```python
# information about data type and count
df1.info()
```

```python
#Check if student ID is unique or not
df1['Student ID'].is_unique
```

```python
# Create a dataset for ela course
df1_ela=df1[['Student ID','Academic Year','District', 'Grade','Race/Ethnicity','ELA Teacher ID','ELA Score','ELA Proficiency Level','School']]
df1_ela
```

```python
#Add a column
df1_ela['Course']='ELA'
```

```python
# Check for total null values in the data
null_mask = df1_ela.isnull().any(axis=1)
null_rows = df1_ela[null_mask]

print(null_rows)
```

```python
# Drop all the NA Values from selected columns
df_ela_final= df1_ela.dropna( how='any',subset=['ELA Teacher ID', 'ELA Score', 'ELA Proficiency Level'])
```

```python
# Rename the column names
df_ela_final=df_ela_final.rename(columns={'ELA Teacher ID':'Teacher ID','ELA Score':'Score','ELA Proficiency Level':'Proficiency Level'})
```

```python
# Creating dataset for Math Course
df1_maths=df1[['Student ID','Academic Year','District', 'Grade','Race/Ethnicity','Math Teacher ID','Math Score','Math Proficiency Level','School']]
# Creating a new column
df1_maths['Course']='Maths'
```

```python
# Chceking Null values
null_mask = df1_maths.isnull().any(axis=1)
null_rows = df1_maths[null_mask]

print(null_rows)
```

```python
#Dropping Null values
df_math_final=df1_maths.dropna(subset=['Math Teacher ID','Math Score','Math Proficiency Level'], how='any')
df_math_final
```

```python
#Renaming the column name
df_math_final=df_math_final.rename(columns={"Math Teacher ID": "Teacher ID", "Math Score": "Score", "Math Proficiency Level": "Proficiency Level"}, errors="raise")
```

```python
#Combining both the ela and math course data
df_final=pd.concat([df_ela_final, df_math_final], axis=0)
```

```python
# Sort the final data by Student ID
df_final=df_final.sort_values(by='Student ID', ascending=True)
df_final
```

```python
# Arranging columns
df_final=df_final[['Student ID', 'Academic Year', 'District', 'Grade', 'Course', 'Race/Ethnicity',
       'Teacher ID', 'Score', 'Proficiency Level', 'School']]
df_final
```
