```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
data_path="/content/drive/MyDrive/geospatil vizuals /Public_School_Characteristics_-_Current.csv"
shape_path="/content/drive/MyDrive/geospatil vizuals /shape file/Public_School_Characteristics_-_Current.shp"
district_shape_path='/content/drive/MyDrive/geospatil vizuals /district shape file/School_District_Boundaries_-_Current.shp'
```

```python
df=pd.read_csv(data_path,low_memory=False)
columns=['LEA_NAME','NMCNTY','LEAID','LEA_NAME','SCH_NAME','LSTATE','LCITY','STUTERATIO']
df=df[columns]
df.columns=['edu_agency name','county','GEOID','edu_agency_name','school_name','location_state','location_city','student_teacher_ratio']

df.dropna()
df_state=df.groupby(['location_state']).mean(['student_teacher_ratio']).reset_index()

df_state
```

```python
import plotly.express as px
fig = px.choropleth(df_state,
                    locations='location_state',
                    locationmode="USA-states",

                    scope="usa",
                    color='student_teacher_ratio',
                    color_continuous_scale="Viridis_r",

                    )
fig.update_layout(
      title_text = 'Student Teacher Ratio by U.S States',
      title_font_family="Times New Roman",
      title_font_size = 22,
      title_font_color="black",
      title_x=0.45,
         )

fig.show()
```

District shape is the shape file for creating boundaries at county level.

```python
district_shape = gpd.read_file(district_shape_path)
```

```python
us_fips=pd.read_csv('/content/drive/MyDrive/geospatil vizuals /us-state-ansi-fips.csv')
us_fips.head(2)
```

```python
district_shape[['STATEFP']]=district_shape[['STATEFP']].astype(int)
```

```python
district_shape_merged=district_shape.merge(us_fips,right_on=" st",left_on="STATEFP",how="inner")
district_shape_merged
```

```python
columns=[' st', ' stusps', 'NAME', 'GEOID', 'geometry']
district_shape_merged=district_shape_merged[columns]
district_shape_merged=district_shape_merged.rename(columns ={' st':'state_fips', ' stusps':'states_abb'})
```

```python
district_shape_merged
```

Creating map for North Corilina

```python
nc_shape=district_shape_merged[district_shape_merged.state_fips==37]
nc_shape
```

```python
df_nc=df[df.location_state=='NC']
df_nc=df_nc.groupby(['GEOID']).mean(['student_teacher_ratio']).reset_index()
```

```python
duplicate = df_nc[df_nc['GEOID'].duplicated()]
duplicate
```

```python
nc_shape[['GEOID']]=nc_shape[['GEOID']].astype(int)
```

```python
nc_shape.info()
```

```python
final_nc_df = nc_shape.merge(df_nc, left_on='GEOID',right_on='GEOID',how='outer')
```

```python
final_nc_df
```

```python
final_nc_df2 = final_nc_df.dropna(subset=['geometry'])
```

```python
final_nc_df2.info()
```

```python
final_nc_df2=final_nc_df2[['NAME','geometry','student_teacher_ratio']]
```

```python
fig = px.choropleth(final_nc_df2, geojson=final_nc_df2.geometry,
                    locations="NAME", color="student_teacher_ratio",
                    height=500,
                   color_continuous_scale="Viridis")
fig.update_geos(fitbounds="locations", visible=True)
fig.update_layout(
    title_text='Map'
)
fig.update_layout(
      title_text = 'Student Teacher Ratio by North Corilina County',
      title_font_family="Times New Roman",
      title_font_size = 22,
      title_font_color="black",
      title_x=0.45,
         )

fig.show()

```
